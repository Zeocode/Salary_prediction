```python
import pandas as pd
df=pd.read_csv("employee_data.csv")

print(df)
```

```python
df1=df[0:100].copy()
del df1['id']
del df1['groups']
del df1['salary']
#don't include salary too: Reason : Kmeans predictions requires all the features
#clustering has been done using all the relevant features except salary.
#Salary feature has been used in the regression
df1
```

```python
from sklearn.cluster import KMeans
di=0.0
di1=0.0

for i in range(2,50):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df1)
    labels=kmeans.predict(df1)
    labels=list(labels)
    centroids=kmeans.cluster_centers_
    #print(centroids)
    min_=centroids[1]-centroids[0]
    min_squared=min_[0]**2+min_[1]**2+min_[2]**2
    min_squared=min_squared**.5
    #print(min_squared)
    #dunn's Index
    for j in range(len(centroids)):
        for k in range(j+1,len(centroids)):
            dist=centroids[k]-centroids[j]
            dist_squared=dist[0]**2+dist[1]**2+dist[2]**2
            dist_squared=dist_squared**.5
            #print(dist_squared)
            if(dist_squared<min_squared):
                min_squared=dist_squared
    print(i)
    sum_=[]
    for j in range(len(centroids)):
        sum_.append(0)
    for j in range(len(labels)):
        for k in range(j+1,len(labels)):
            if(labels[j]==labels[k]):
                dist1=(df1.iloc[j]['age']-df1.iloc[j]['age'])**2+(df1.iloc[j]['healthy_eating']-df.iloc[k]['healthy_eating'])**2+(df1.iloc[j]['active_lifestyle']-df1.iloc[j]['active_lifestyle'])**2
                #print(dist1)
                dist1_squared=dist1**.5
                #print(j," ",k)
                sum_[labels[j]]=sum_[labels[j]]+dist1_squared
    #print("check")               
    for j in range(len(centroids)):
        cn=labels.count(j)
        sum_[j]=(sum_[j])/(cn) # cannot allow (cn-1) as division by zero may occur
            
    #print(type(sum_))  
    #print(sum_)
    di=min_squared/max(sum_)
    print(di)
    #print(i)
    if (di>di1):
        di1=di
        final_n=i
        continue
    else:
        break
print("Optimal Number of Clusters = ",final_n)
   
kmeans = KMeans(n_clusters=final_n)
kmeans.fit(df1[0:90])
labels=kmeans.predict(df1[90:100])   
labels1=kmeans.predict(df1[0:90])

```
