```python
import pandas as pd
df=pd.read_csv("employee_data.csv")

print(df)
```

```python
df1=df.copy()
del df1['id']
del df1['groups']
del df1['salary']
#don't include salary too: Reason : Kmeans predictions requires all the features
#clustering has been done using all the relevant features except salary.
#Salary feature has been used in the regression
df
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5)
kmeans.fit(df1[0:900])
labels=kmeans.predict(df1[900:1000])

labels1=kmeans.predict(df1[0:900])

```

```python
df2=pd.DataFrame()
df2['Labels']=labels
df2

```

```python
train_df=df[0:900].copy()
train_df['Labels']=labels1
train_df
```

```python
test_df=df[900:1000].copy()
test_df['Labels']=labels
test_df
```

```python
sal_list=[]

sal=[]
from sklearn import linear_model
reg=linear_model.LinearRegression()

for i in range(0,100):
    j=train_df.iloc[i,6]    
    cluster=train_df[train_df.Labels==j]
    reg.fit(cluster[['age','healthy_eating']],cluster.salary)
    sal_list = reg.predict(test_df[i:i+1][['age','healthy_eating']])
    
    sal.append(sal_list[0])
    
sal
```

```python
test_df['pred_salary']=sal
test_df
```
